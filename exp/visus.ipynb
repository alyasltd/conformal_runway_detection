#### **Visualisation** 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
```

```python
df_results_v5a = pd.read_csv("results_v5a.csv")
df_results_v5m = pd.read_csv("results_v5m.csv")
df_results_v6a = pd.read_csv("results_v6_a.csv")
df_results_v6m = pd.read_csv("results_v6_m.csv")

df_results_v5a_synth = pd.read_csv("results_v5_synth_a.csv")
df_results_v6a_synth = pd.read_csv("results_v6_synth_a.csv")
df_results_v5m_synth = pd.read_csv("results_v5_synth_m.csv")
df_results_v6m_synth = pd.read_csv("results_v6_synth_m.csv")
```

```python
sns.set_style("whitegrid")
plt.rcParams.update({
    "font.size": 12,
    "axes.titlesize": 16,
    "axes.labelsize": 14,
    "legend.fontsize": 12
})

palette = {
    "YOLOv6": "#340cb2",  # violet prune
    "YOLOv5": "#D35400",  # orange brûlé
}

```

##### **IOU pred VS IOU pred C**

```python
# --- IoU pred vs IoU pred C (global - additive) ---
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_results_v5a, x="iou_pred_gt", y="iou_pred_c_gt",
                marker="*", s=150, color=palette["YOLOv5"], alpha=0.7, label="YOLOv5 - Additive")
sns.scatterplot(data=df_results_v6a, x="iou_pred_gt", y="iou_pred_c_gt",
                marker="^", s=100, color=palette["YOLOv6"], alpha=0.7, label="YOLOv6 - Additive")
plt.xlabel("YOLO Prediction IoU")
plt.ylabel("IoU after conformal prediction")
#plt.title("IoU pred vs IoU pred C (Global - Additive)")
plt.legend()
plt.tight_layout()
plt.savefig("/home/aws_install/conformal_runway_detection/plots/iou_pred_vs_iou_pred_c_additive.png", dpi=300)
plt.show()

# --- IoU pred vs IoU pred C (global - multiplicative) ---
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_results_v5m, x="iou_pred_gt", y="iou_pred_c_gt",
                marker="*", s=150, color=palette["YOLOv5"], alpha=0.7, label="YOLOv5 - Multiplicative")
sns.scatterplot(data=df_results_v6m, x="iou_pred_gt", y="iou_pred_c_gt",
                marker="^", s=100, color=palette["YOLOv6"], alpha=0.7, label="YOLOv6 - Multiplicative")
plt.xlabel("YOLO Prediction IoU")
plt.ylabel("IoU after conformal prediction")
#plt.title("IoU pred vs IoU pred C (Global - Multiplicative)")
plt.legend()
plt.tight_layout()
plt.savefig("/home/aws_install/conformal_runway_detection/plots/iou_pred_vs_iou_pred_c_multiplicative.png", dpi=300)
plt.show()
```
